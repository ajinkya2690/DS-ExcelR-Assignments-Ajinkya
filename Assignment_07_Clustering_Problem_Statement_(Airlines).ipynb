```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
import scipy.cluster.hierarchy as sch
```

```python
Data = pd.read_excel("/content/EastWestAirlines.xlsx", sheet_name = "data")
Data
```

```python
Data.info()
```

```python
Data1 = Data.drop("ID#", 1)
Data1
```

```python
Data2 = Data1.copy()
```

```python
Norm_Data = pd.DataFrame(normalize(Data2),columns = Data2.columns)
Norm_Data
```

Now, the data is Normalize. So, we start Clustring


# K-Means Clustring

```python
WCSS = []
for i in range(1,11):
    Kmeans = KMeans(n_clusters = i, random_state = 48)
    Kmeans.fit(Norm_Data)
    WCSS.append(Kmeans.inertia_)
WCSS
```

```python
#Elbow Plot
plt.plot(range(1,11),WCSS)
plt.title("Elbow Plot")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.xticks(range(1,11))
plt.show()
```

```python
Clusters = KMeans(4,random_state = 48)
Clusters.fit(Norm_Data)
Clusters.labels_
```

```python
Data2["Cluster"] = Clusters.labels_
Data2
```

```python
Data2.groupby("Cluster").agg(["mean"]).reset_index()
```

```python
plt.scatter(Data2["Cluster"],Data2["Balance"],c = Clusters.labels_)
```

```python
Clusters.inertia_
```

```python
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.ylabel('Distance')
dendrogram = sch.dendrogram(sch.linkage(Norm_Data, method='complete',metric='euclidean'))
```

```python
Data3 = Data1.copy()
```

```python
HC = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "complete")
HC.fit(Norm_Data)
HC.labels_
```

```python
Data3["Cluster"] = HC.labels_
Data3
```

```python
Data3.groupby("Cluster").agg("mean").reset_index()
```

```python
plt.scatter(Data3["Cluster"],Data3["Balance"],c = HC.labels_)
```
