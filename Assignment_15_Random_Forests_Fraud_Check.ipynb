```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
Data = pd.read_csv("/content/Fraud_check.csv")
Data.head()
```

```python
Data.info()
```

```python
Data.describe()
```

```python
Data1=pd.Series(Data['Taxable.Income'])
s=[]
for i in Data1:
    if i>30000:
        s.append('Risky')
    else:
        s.append('Good')
print(s)
```

```python
Data2=pd.DataFrame(s)
Data3=pd.concat([Data2,Data],axis=1)
Data3=Data3.rename({0:'o/p'},axis=1)
Data3
```

```python
labelencoder=LabelEncoder()
Data3.iloc[:,0]=labelencoder.fit_transform(Data3.iloc[:,0])
Data3.iloc[:,1]=labelencoder.fit_transform(Data3.iloc[:,1])
Data3.iloc[:,2]=labelencoder.fit_transform(Data3.iloc[:,2])
Data3.iloc[:,-1]=labelencoder.fit_transform(Data3.iloc[:,-1])
```

```python
Data3
```

```python
array=Data3.values
```

```python
X = array[:,2:10]
Y = array[:,0]
```

```python
num_trees = 100
max_features = 5
kfold = KFold(n_splits=10)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```
